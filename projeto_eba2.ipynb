### 1.1.1. Dataset 'PAG_HEI':

Esse dataset fornece informações sobre a qualidade da alimentação de cada indivíduo com informações do Healthy Eating Index (HEI), que é um índice desenvolvido pela USDA para avaliar a qualidade da dieta com base em diretrizes recomendadas. Além disso, contém informações sobre a atividade física de cada pessoa.

* **SEQN:** Identificador único de cada pessoa
* **PAG_MINW:** Tempo total semanal de atividade aeróbica moderada-vigorosa (mim)
* **ADHERENCE:** Grupo de aderência ao PAG (1 = baixo (<150 min/semana), 2 = Adequado (150 ~ 300) e 3 = Acima (>300))

As próximas colunas fazem parte de um conjunto de informações sobre o índice HEI de cada paciente. O índice tenta refletir o grau em que a dieta adere aos padrões alimentares saudáveis, sendo que quanto mais alto o valor, maior aderência aos padrões recomendados. Esse conjunto de variáveis, possui valores que variam de 0 a 5 para uma série de componentes como: frutas, proteínas, laticínios e etc. Algumas outras variáveis como Gordura Saturada, Açucares, variam de 0 a 10. Por último, temos o  Total Score que soma os valores de cada uma das variáveis.

* **HEI2015C1_TOTALVEG:** HEI - Total de vegetais 
* **HEI2015C2_GREEN_AND_BEAN**: HEI - Verduras verde-escuras e leguminosas
* **HEI2015C3_:TOTALFRUIT** HEI - Frutas totais
* **HEI2015C4_:WHOLEFRUIT** HEI - Frutas in natura
* **HEI2015C5_:WHOLEGRAIN** HEI - Grãos naturias
* **HEI2015C6_:TOTALDAIRY** HEI - Laticínios
* **HEI2015C7_:TOTPORT** HEI - Proteínas
* **HEI2015C8_:SEAPLANT_PROT** HEI - Plantas marinhas e proteínas vegetais
* **HEI2015C9_:FATTYACID** HEI - Ácido-graxo
* **HEI2015C10_:SODIUM** HEI - Sódio
* **HEI2015C11_:REFINEDGRAIN** HEI - Grãos Refinados
* **HEI2015C12_:SFAT** HEI - Gordura Saturada
* **HEI2015C13_:ADDSUG** HEI - Açucares adicionados
* **HEI2015_TOTAL_SCORE:** HEI - Escore total (0 a 100), onde quanto mais próximo do 100, melhor a dieta.

### 1.1.2. Dataset 'DEMO_PHQ':

Este dataset contém informações com base no Patient Health Questionnaire - 9 (PHQ-9). O PHQ-9 é uma ferramenta usada para avaliar a presença e a gravidade de sintomas de depressão em pacientes. O questionário é composto por 9 perguntas que avaliam sintomas presentes nas últimas semanas. Os pacientes devem classificar cada pergunta conforme a frequência com que determinado sintoma foi presente, sendo: 
* 0: Nunca 
* 1: Vários dias
* 2: Mais da metade dos dias
* 3: Quase todos os dias

Os valores 7 e 9 representam "Se recusou a responder" e "Não sabe", respectivamente. 

A lista de sintomas em cada pergunta é:

* **DPQ010:** Pouco interesse ou pouco prazer em fazer as coisas
* **DPQ020:** Se sentiu para baixo, deprimido(a) ou sem perspectiva
* **DPQ030:** Dificuldade para pegor no sono, permanecer domindo ou dormir mais do que o costume
* **DPQ040:** Se sentiu cansado(a) ou com pouca energia
* **DPQ050:** Falta de apetitve ou comeu demais
* **DPQ060:** Se sentiu mal consido mesmo(a) ou achou que é um fracasso ou que decepcionou sua família ou a você mesmo(a)
* **DPQ070:** Dificuldade para se concentrar nas coisas (como ler o jornal ou ver televisão)
* **DPQ080:** Teve lentidão para se movimentar ou falar (a ponto de outras pessoas perceberem), ou ao contrário, esteve tão agitado (a) que você ficada andando de um lado para outro mais do que costume
* **DPQ090:** Pensou em se ferir de alguma maneira ou que seria melhor estar morto(a)

Além dos dados do questionário, o dataset também inclui uma série de dados demográficos dos pacientes:

* **RIAGENDR:** Gênero (1: Masculino, 2: Feminino)
* **RIDAGEYR:** Idade em anos
* **RIDRETH1:** Raça/Etnia (1 = Branco não-hispânico, 2 = Negro não-hispânico, 3 = Mexicano-americano, 4 - Outro, 5 = Outro - Hispânico)
* **DMDEDUC:** Escolaridade (1 = Menor que 9o ano, 2 = 9~12o ano, 3 = Ensino Médio, 4 = Ensino Superior Incompleto, 5 = Superior Completo, 7 = Se recusou a responder e 9 = Não sabe)
* **INDFMINC:** Renda anual familiar (US$)

```python
# Função para investigar a presença de valores nulos em cada uma das características em quantidade e porcentagem:
def show_null(df):
    '''
    Input: Dataframe
    Output: Dataframe onde cada linha é uma coluna do input e cria as colunas:
        Quantidade: Contagem da quantida de nulos presente naquela coluna
        % de nulos: Expressa a quantidade de nulos em porcentagem
        Tipo: Informa o tipo de variável da coluna
    '''
    null_columns = (df.isnull().sum(axis=0)/len(df)*100).sort_values(ascending=False).index
    
    null_data = pd.concat([df.isnull().sum(axis=0), 
                            (df.isnull().sum(axis=0)/len(df)*100).sort_values(ascending=False),
                            df.loc[:,df.columns.isin(list(null_columns))].dtypes], 
                            axis = 1)
    
    null_data = null_data.rename(columns = {0:'Quantidade', 1:'% de nulos', 2:'Tipo'}).sort_values(ascending=False, by = '% de nulos')
    return null_data

show_null(df)
```

### 1.3. Análise inicial em relação ao questionário de sintomas de depressão:

Podemos ver que existem uma grande quantidade de 0 (zeros), o que significa que os individuos alegam nunca terem apresentados os sintomas. Com excessão das perguntas DPQ030 e DPQ040, pelo menos 75% das respostas é 0 (zero). 

Em relação a média e desvio-padrão, não podemos tirar muitas conclusões pois o valor 9 cria uma distorção desses dados, já que o 9 representa o "Não sabe", porém ele impacta diretamente nas medidas nas medidas como média, máximo, desvio-padrão e etc.

Vamos tratar os dados ausentes e 9 ('Não sabe'), da mesma forma. 

```python
# Lista dos possíveis valores de resposta no questionário dos sintomas de depressão:
valores_sintomas = [0,1,2,3,9,np.NaN]

# Cria uma tabela com a contagem de cada valor em cada coluna do questionário:
tabela_contagem_sintomas = pd.DataFrame(index=valores_sintomas, columns = quest_sintomas)

for coluna in quest_sintomas:
    # Cria uma coluna na tabela fazendo um value_counts na coluna e indexando de maneira padrão:
    tabela_contagem_sintomas[coluna] = df[coluna].value_counts(dropna = False).reindex(valores_sintomas)

# Mostra a contagem de valores por coluna
tabela_contagem_sintomas
```

#### Frequência de respostas para cada resposta:

Já podemos perceber que existem alguns sintomas que possuem uma frequência de respostas positivas maiores em relações a outros. Como é o caso do DPQ030 e DPQ040, que possuem maiores quantidades de 1 ('Varios dias'), 2 ('Mais da metade dos dias') e 3 ('Quase todos os dias').

As perguntas em questão eram:

* **DPQ030: Dificuldade para pegar no sono, permanecer dormindo ou dormir mais do que o costume**
* **DPQ040: Se sentiu cansado(a) ou com pouca energia**

Inclusive, são perguntas que são relativamente conectadas, o fato de uma pessoa dormir mal poderia levar a um cansado maior durante o dia.

```python
# Transforma as respostas 7 ('Não respondeu') 9 ('Não sabe') em valor nulo. Isso serve para não interferir nas operações e medidas:
df[quest_sintomas] = df[quest_sintomas].replace([7,9],np.NaN)
# Coluna Escolaridade e Renda Anual, junta as respostas 7 e 9 em uma só.
df['DMDEDUC'] = df['DMDEDUC'].replace([7,9],9)
df['INDFMINC'] = df['INDFMINC'].replace([77,99],99)

# Encontrar as linhas que possuem 23 ou mais valores nulos
linhas_mais_nulos = df[df.isnull().sum(axis=1) >= 23]

# Quantidade de linhas com, pelo menos, 23 valores nulos.
print(f'Quantidade de linhas com muitos valores ausentes: {linhas_mais_nulos.shape[0]}')

# Excluindo as linhas que não possuirem, pelo menos, 8 colunas preenchidas.
df.dropna(thresh=df.shape[1] - 22, inplace = True)
print(f'Shape do Dataset resultante: {df.shape}')
```

#### Tratamentos:

* Tratamos os valores 9 como sendo nulos, já que não temos a resposta. 
* Encontramos as colunas que uma grande deficiencia de dados (mais de 23 valores nulos) e chegamos ao valor de 175 linhas. 
* Excluímos essas 175 linhas por possuírem pouquíssima informação. 

#### Criação de novas colunas: 

* Vamos criar a coluna **phq9** que irá somar os valores de cada uma das respostas do questionário de sintomas de depressão. 
* Vamos criar a coluna **phq_grp** que irá segmentar a coluna anterior em 3 classes: 'Sem sintomas' (phq menor ou igual a 5), 'Sintomas leves' (phq entre 6 e 10) e 'Sintomas moderados e severos' (phq maior ou igual a 10)


```python
# Cria a coluna 'phq9' que corresponde ao escore total do questionário 'PHQ-9': 
# phq9 = DPQ010 + DPQ020 + ... + DPQ090

df['phq9'] = df[quest_sintomas].sum(axis=1)

# Cria uma nova coluna com as classificações:
condicoes = [(df['phq9'] < 5), ((df['phq9'] >= 5) & (df['phq9'] < 10)), (df['phq9'] >= 10)]
classificacoes = ['sem sintomas', 'sintomas leves', 'sintomas moderados e severos']
df['phq_grp'] = np.select(condicoes, classificacoes, default = np.nan)
```

# 4. Testes de Hipóteses:

Ao longo da nossa análise de dados, pudemos notar algumas relações entre variáveis, porém, não queremos fazer afirmações sem uma boa "justificativa" por trás. Os testes de hipóteses são uma ferramente chave para, a um certo nível de confiança, fazer alguns apontamentos sobre o nosso conjunto de dados. Sendo assim, vamos testar as seguintes questões:

* **4.1 O escore total da alimentação é uma distribuição normal**
* **4.2 Em relação aos sintomas de depressão, existe uma diferença significativa entre as mulheres e os homens?**
* **4.3 Existe uma relação entre os sintomas de depressão e a atividade física?**
* **4.4 Existe uma relação entre sintomas de depressão e a qualidade da alimentação?**
* **4.5 Existe uma relação entre sintomas de depressão e a escolaridade dos indivíduos?**
* **4.6 Existe uma relação entre sintomas de depressão e as classes de renda dos indivíduos?**
* **4.7 Existe uma relação entre sintomas de depressão e a idade dos indivíduos?**

### Antes disso, vamos entender o que é um teste de hipótese:

Teste de hipótese é um elemento estatístico que é usado para tomar decisões sobre uma afirmação (hipótese) em relação aos dados de uma amostra. A estrutura básica de um teste de hipótese envolve duas hipóteses:

* **Hipótese Nula ou H0:** É a afirmação que queremos testar, geralmente afirmando que não há efeito ou relação entre as variáveis na população. Por exemplo, H0 poderia ser: As médias no questionário da alimentação dos homens e das mulheres são iguais.
* **Hipótese Alternativa ou H1:** É a afirmação aposta à hipótese nula, é a hipótese que queremos aceitar se houver uma diferença ou efeito entre as variáveis. Em relação ao mesmo exemplo, seria: A média dos homens é diferente da média das mulheres em relação ao questionário de alimentação. 

Com isso, calculamos uma estatística de teste com base nos dados da amostra e comparamos com uma distribuição de probabilidade conhecida. O resultado dessa comparação indica se devemos rejeitar ou não a hipótese nula. Expressamos o resultado como um p-valor, que é, basicamente, a probabilidade de se observar os dados daquela amostra assumindo que a hipótese nula seja verdadeira. Ou seja, o p-valor indica se o que temos é algo muito raro ou difícil de acontecer, considerante que a hipótese nula seja verdadeira. Compara-se o p-valor encontrado com o nível de significância do teste, e isso indica se devemos rejeitar ou não a hipótese nula. O nível de significância (alpha) é o complemento do nível de confiança do teste, em relação a 1. O nível de confiança é um valor arbitrário que devemos definir para o nosso teste, onde geralmente opta-se por valores como 0.95, 0.99 ou até mais. 

Em geral, se p-valor < alpha, rejeita-se a hipótese nula. E se p-valor > alpha, não rejeitamos a hipótese nula. 

Existem vários tipos de estatísticas de teste que informam um p-valor. O teste que melhor se adequa ao seu problema é uma situação que precisa ser analisada e estudada. Essa escolha depende muito dos dados que você possui. 

## 4.1 O escore total da alimentação é uma distribuição normal?

Como vimos anteriormente na análise univariada do escore total do questionário da alimentação, aparentemente a distribuição poderia ser próxima de uma normal. Para isso, vamos usar uma técnica estatística para determinar se duas amostras provêm de uma mesma distribuição, e neste caso, queremos descobrir se o escore total vem de uma distribuição normal. 

É importante realizar um teste de normalidade porque alguns testes de hipótese requerem que os dados sigam uma distribuição normal. 

O nome do teste que iremos fazer é **Teste de Kolgomorov-Smirnov**. As hipóteses do Teste KS são:
* Hipótese Nula (H0): A amostra é proveniente de uma distribuição normal
* Hipótese Alternativa (H1): As duas distribuições são diferentes

```python
# Selecionando a coluna do Teste de Hipótese:
data = df['HEI2015_TOTAL_SCORE'].dropna()
# Obtendo média e desvio padrão dos dados para criar a normal adaptada:
loc, scale = norm.fit(data)
# Criando uma distribuição normal de mesma média e desvio padrão:
n = norm(loc=loc, scale=scale)
# Realizando o teste da amostra em relação a normal:
stats, p = kstest(data, n.cdf)

print("Kolmogorov-Smirnov Teste:")
print("stat    = %9.5f" % stats)
print("p-valor = %9.5f" % p)
```

Realizando o **Teste de Kolgomorov-Smirnov**, obtivemos o **p-valor de 0.00322**, que é menor do que 0.05 (alpha) e portando devemos rejeitar a hipótese nula. Sendo assim, **não podemos afirmar que a distribuição do escore total é uma distribuição normal.** 

## 4.2 Em relação aos sintomas de depressão, existe uma diferença significativa entre as mulheres e os homens?

Para isso, vamos utilizar um outro teste de hipótese. Nesse caso, **queremos comparar duas amostras e verificar se existe uma diferença significativa entre elas**. Vamos separar os homens das mulheres e olhar para o sintomas de depressão, segundo a variável **phq9**, que informa a quantidade de sintomas observados pelo indivíduo.

Utilizaremos o **Teste t de Student**, que terá as seguintes hipóteses:
* Hipótese Nula (H0): Não há diferença entre as médias das duas amostras.
* Hipótese Alternativa (H1): Existe uma diferença entre as médias das amostras.


```python
# Separar homens das mulheres em dois dataframes diferentes
df_homens = df.dropna().loc[df['RIAGENDR'] == 1].reset_index()
df_mulheres = df.dropna().loc[df['RIAGENDR'] == 2].reset_index()

# Teste t:
stats, p = ttest_ind(df_homens['phq9'], df_mulheres['phq9'])
print("Teste t de Student:")
print("stat    = %9.5f" % stats)
print("p-valor = %9.5f" % p)
```

## 4.3 Existe uma relação entre os sintomas de depressão e a atividade física?

* **2. Vamos fazer um teste de hipótese com 3 amostras (ANOVA), onde vamos analisar a variável "phq9" em cada uma das classes de "ADHERENCE", onde classifica em 'Baixo', 'Adequado' e 'Acima' para a atividade fisica semanal.**

Hipóteses do teste ANOVA (3 fatores):

* Hipótese Nula (H0): As médias de todas as amostras são iguais.
* Hipótese Alternativa (H1): Pelo menos uma das médias das amostras é diferente das outras



```python
# Correlação (Coeficiente de Pearson):
corr_pearson = df['PAG_MINW'].corr(df['phq9'])
corr_spearman, p_spearman = spearmanr(df['PAG_MINW'], df['phq9'])
print('Correlação entre "PAG_MINW" e "phq9":')
print(f'Coeficiente de Pearson: {corr_pearson.round(4)}')
print(f'Coeficiente de Spearman: {corr_spearman.round(4)}')
print(f'P-valor de Spearman: {p_spearman.round(6)}')
```

```python
# Dividir as classes:
df_exec_baixo = df.dropna().loc[df['ADHERENCE'] == 1]
df_exec_adequado = df.dropna().loc[df['ADHERENCE'] == 2]
df_exec_acima = df.dropna().loc[df['ADHERENCE'] == 3]

# Teste ANOVA (3 fatores):
print('Teste ANOVA - 3 fatores:')
stats, p = f_oneway(df_exec_baixo['phq9'], df_exec_adequado['phq9'], df_exec_acima['phq9'])
print("stat    = %9.5f" % stats)
print("p-valor = %9.5f" % p)
```

O sinal de negativo indica que as variáveis tem sentidos opostos, ou seja, conforme uma cresce, a outra tende a decrescer. Porém, como o valor da correlação é muito baixo, não vamos adentrar nesse aspecto da correlação. 

Em relação ao **Teste ANOVA** dos 3 grupos de ADHERENCE em relação a variável do questionário de sintomas de depressão, o **teste de hipótese indica que devemos rejeitar a hipótese nula**. Sendo assim, a hipótese alternativa nos sugere que as médias não são iguais, com 95% de confiança.

O teste executado possui a limitação de não nos indicar qual das médias é diferente das demais. Sendo assim, vamos aplicar um teste t de Student em dois pares de classes. 

* **Teste 1:** Verificar se existe uma diferença estatísticamente significativa entre as classes de "Baixo" e "Adequado" (ADHERENCE) em relação a variável "phq9" (sintomas de depressão).

* **Teste 2:** Semelhante ao caso anterior, mas vamos analisar as classes "Adequado" e "Acima". 

As hipóteses dos testes são:
* Hipótese Nula (H0): Não há diferença entre as médias das duas amostras.
* Hipótese Alternativa (H1): Existe uma diferença entre as médias das amostras.

```python
# Teste 1:
stats, p = ttest_ind(df_exec_baixo['phq9'],  df_exec_adequado['phq9'])
print("Teste 1 (t de Student) - Baixo x Adequado:")
print("stat    = %9.5f" % stats)
print("p-valor = %9.5f" % p)
print("-------------------------------------------")
# Teste 1:
stats, p = ttest_ind(df_exec_acima['phq9'],  df_exec_adequado['phq9'])
print("Teste 2 (t de Student) - Adequado x Acima:")
print("stat    = %9.5f" % stats)
print("p-valor = %9.5f" % p)
```

No **Teste 1: Baixo x Adequado**, tivemos um p-valor < 0.05, ou seja, **rejeita-se a hipótese nula que dizia que as médias eram iguais**. Sendo assim, sugere-se que **exista uma diferença estatísticamente significativa entre o grupo que pratica atividade física abaixo do recomendado e o grupo que pratica de forma adequada**, em relação a variável 'phq9', que é o questionário de sintomas de depressão. Já em relação ao **Teste 2: Adequado x Acima**, o p-valor foi de 0.17, ou seja, não devemos rejeitar a hipótese nula. Sendo assim, **não existe uma diferença significativa entre quem faz atividade física recomendada e quem faz acima do recomendado**. 

Com isso, podemos ver que **existe sim uma relação entre a atividade física e os sintomas de depressão. As pessoas mais ativas possuem um quadro melhor em relação a aparição desses sintomas**. Além disso, fica evidente a importância da atividade física recomendada, haja vista a enorme quantidade de pesquisas que indicam tantos outros benefícios dessa atividade, como melhor do sistema respiratório, cardiovascular, pressão arterial, níveis de gordura e etc.

Dada a complexidade do ser humano, dificilmente um quadro depressivo conseguiria ser explicado por poucos e/ou pequenos fatores. Entendemos que a alimentação e a atividade física podem ser elementos importantes para a avaliação do problema, mas existem outros elementos que poderiam afetar o quadro observado. Seria interessante obter mais dados sobre os indivíduos, como por exemplo: dados sobre a percepção da sua vida social, a presença de algum trauma recente como separação, perda familiar ou demissão de emprego.
